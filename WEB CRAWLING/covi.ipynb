```python
import requests
import csv
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python


# URL của trang web chứa dữ liệu COVID-19
URL = "https://www.worldometers.info/coronavirus/#countries"

# Gửi yêu cầu HTTP để lấy nội dung trang web
response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")

# Tìm bảng chứa dữ liệu
table = soup.find("table", id="main_table_countries_today")
headers = []
for th in table.find_all("th")[1:]:
    aria_label = th.get("aria-label")
    if aria_label != "#":
         headers.append(th.text.strip())
    else:
        print(f"Bỏ qua cột với aria-label: {aria_label}")

print(headers)
```

```python



# Lấy dữ liệu từ các hàng trong bảng
data = []
for row in table.find("tbody").find_all("tr")[1:]:
    cells = row.find_all("td")
    # Kiểm tra nếu hàng có chứa liên kết (tức là hàng của quốc gia)
    if cells[1].find("a"):
        row_data = [cell.text.strip().replace(',', '') for cell in cells]
        if row_data:
            data.append(row_data)
#Lưu dữ liệu vào file CSV
filename = "covid_data.csv"
with open(filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(headers)  # Ghi tiêu đề
    writer.writerows(data)    # Ghi dữ liệu

print(f"Dữ liệu đã được lưu vào {filename}")
```

```python
# Replace NaN values with 0 and convert columns to integers
df['TotalRecovered'] = df['TotalRecovered'].fillna(0).astype(int)
df['TotalCases'] = df['TotalCases'].fillna(0).astype(int)

# Calculate the recovery rate
df['RecoveryRate'] = (df['TotalRecovered'] / df['TotalCases'] * 100)

df
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://web.archive.org/web/20220324004745/https://www.worldometers.info/coronavirus/#countries'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Tìm bảng với id 'main_table_countries_today'
table = soup.find('table', id='main_table_countries_today')

# Lấy hàng tiêu đề để xác định cột
header_row = table.find('thead').find_all('tr')[0]  # Lấy từ <thead> để chắc chắn đúng
columns = [cell.text.strip() for cell in header_row.find_all('th')]

# Kiểm tra vị trí của cột "New Cases"
index = 3  # Cột "New Cases" (có thể thay đổi theo trang web thực tế)

# Danh sách các hàng tổng cần loại bỏ
EXCLUDE_WORDS = {"World", "Total:", "All Countries", ""}

# Trích xuất dữ liệu cho từng quốc gia
data = []
for row in table.find_all('tr')[9:]:  # Bỏ qua các dòng không cần thiết
    cells = row.find_all('td')
    
    if len(cells) > index:
        country = cells[1].text.strip()  # Lấy tên quốc gia

        # Nếu quốc gia thuộc danh sách loại bỏ, thì bỏ qua
        if country in EXCLUDE_WORDS:
            continue

        new_cases_text = cells[index].text.strip()

        # Xóa dấu "+" và dấu phẩy để chuyển đổi thành số
        new_cases_text = re.sub(r'[\+,]', '', new_cases_text)
        try:
            new_cases = int(new_cases_text) if new_cases_text else 0
        except ValueError:
            new_cases = 0  # Nếu vẫn lỗi, đặt về 0

        data.append((country, new_cases))

data
```

```python
# top 5 quốc gia có số ca nhiễm mới cao nhất
top_5_new_cases = sorted(data, key=lambda x: x[1], reverse=True)[:5]
top_5_new_cases
```
